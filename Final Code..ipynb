```python
import pandas as pd
import plotly.graph_objs as go


bitcoin_df = pd.read_csv("C:/Users/krort/Downloads/INFO450/Bitcoin prices.csv")
ethereum_df = pd.read_csv("C:/Users/krort/Downloads/INFO450/Ethereum prices.csv")




#I uppcased the letter D in column date
bitcoin_df = bitcoin_df.rename(columns={"date": "Date"})


ethereum_df = ethereum_df.rename(columns={"date": "Date"})


#I created an empty plotly figure object 
fig = go.Figure()

#For BTC and ETH I was able to create a scatter plot following the trends of the line 

fig.add_trace(
    go.Scatter(x=bitcoin_df['Date'], y=bitcoin_df['Close'], mode='markers', name="Bitcoin",
               line=dict(color='blue'),  
               marker=dict(size=4),  
               )
)

fig.add_trace(
    go.Scatter(x=ethereum_df['Date'], y=ethereum_df['Close'], mode='markers', name="Ethereum",
               line=dict(color='orange'), 
               marker=dict(size=4),  
               )
)

#This was the code I use previously from the trend line and used it to trace both currencies. Gives the average of closing price
#in a 30 window
fig.add_trace(
    go.Scatter(x=bitcoin_df['Date'], y=bitcoin_df['Close'].rolling(window=30).mean(),
               mode='lines', name="Bitcoin Trendline", line=dict(color='blue'))
)

fig.add_trace(
    go.Scatter(x=ethereum_df['Date'], y=ethereum_df['Close'].rolling(window=30).mean(),
               mode='lines', name="Ethereum Trendline", line=dict(color='orange'))
)

#as the other graphs, created and update the chart
fig.update_layout(title="Scatter Plot of BTC and ETH",
                  xaxis_title="Date",
                  yaxis_title="Price (USD)")

fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go

bitcoin_df = pd.read_csv("C:/Users/krort/Downloads/INFO450/Bitcoin prices.csv")
ethereum_df = pd.read_csv("C:/Users/krort/Downloads/INFO450/Ethereum prices.csv")

# Get the closing prices for Bitcoin and Ethereum over the last 30 days
# Create a grid of x and y values that matches the dimensions of the corr matrix
# Calculate the correlation matrix between the two cryptocurrencies
bitcoin_last_30 = bitcoin_df.tail(30)['Close']
ethereum_last_30 = ethereum_df.tail(30)['Close']

corr = np.corrcoef(bitcoin_last_30, ethereum_last_30)

x, y = np.meshgrid(np.arange(2), np.arange(2))

# I tried creating the heatmap.
fig = go.Figure(data=go.Heatmap(
    x=x.flatten(),
    y=y.flatten(),
    z=corr,
    colorscale='Viridis'))

#I have to update the layout of the figure and setting the titles.
fig.update_layout(title="Correlation between Bitcoin and Ethereum Closing Prices over the Last 30 Days",
                  xaxis_title="",
                  yaxis_title="",
                  width=500,
                  height=500)

fig.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

#This is my attempt at creating a Machine Learning model. It takes the data set and you ask a year you want to know the price
#of BTC and it will tell you. You can ask as many times until you say no.

bitcoin_df = pd.read_csv("C:/Users/krort/Downloads/INFO450/Bitcoin prices.csv")

bitcoin_df["Date"] = pd.to_datetime(bitcoin_df["Date"])

#I had to add this so I can take the year and it will take it as a interger because it would not take as a timestamp
bitcoin_df["Year"] = bitcoin_df["Date"].dt.year.astype(int)

#this just gets the closing price and year 
bitcoin_close = bitcoin_df["Close"]
bitcoin_year = bitcoin_df["Year"]

#linear regression model
model = LinearRegression()

#this is the training ml model using the fit method using LR model. Fitting a model to the BTC closing prices over the year.
model.fit(bitcoin_year.values.reshape(-1, 1), bitcoin_close)

while True:
   #takes the input of the year the user wants and converts to integer
    input_year = input("What year do you want to know the price of Bitcoin? ")
    input_year = int(input_year) 

    predicted_price = model.predict([[input_year]])

   
    print("The predicted price of Bitcoin for the year", input_year, "is", predicted_price[0])

    
    another_year = input("Do you want to pick another year? (yes/no) ")

    if another_year.lower() == "no":
        break
```
